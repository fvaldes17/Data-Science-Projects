```python
from EMNIST import x_train, y_train, x_test, y_test
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from matplotlib import pyplot as plt
from random import randint, shuffle
import numpy as np
import pandas as pd
import time
import database_creada
```

```python
x_train_gnb = list(x_train)
y_train_gnb = y_train
x_test_gnb = list(x_test)
y_test_gnb = y_test
```

```python
clf_gnb = GaussianNB()
```

```python
clf_gnb.fit(x_train_gnb, y_train_gnb)
```

```python
predicted_gnb = clf_gnb.predict(x_test_gnb)
expected_gnb = y_test_gnb
```

```python
x_test_gnb = np.array(x_test_gnb)
```

```python
clases62 = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
fig = plt.figure(figsize=(15, 15))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
for j in range(256):
    i = randint(0, len(predicted_gnb) - 1)
    ax = fig.add_subplot(16, 16, j + 1, xticks=[], yticks=[])
    ax.imshow(x_test_gnb.reshape(-1, 28, 28)[i], cmap=plt.cm.binary, interpolation='nearest')
    if predicted_gnb[i] == expected_gnb[i]:
        ax.text(0, 7, clases62[predicted_gnb[i]], fontsize=20, color='green')
    else:
        ax.text(0, 7, clases62[predicted_gnb[i]], fontsize=20, color='red')
ax.axis("on")
fig.savefig('GNB EMNIST DESBALANCEADO.png', dpi=fig.dpi, edgecolor="purple")
```

```python
print(metrics.classification_report(expected_gnb, predicted_gnb))
```

```python
df = pd.DataFrame(metrics.confusion_matrix(expected_gnb, predicted_gnb))
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df)
```

```python
train_counts = list()
y_train_gnb = list(y_train)
for i in range (62):
    train_counts.append(y_train_gnb.count(i))
for i in range (len(y_train_gnb)):
    y_train_gnb[i] = [y_train_gnb[i], i]
y_train_gnb.sort()
x_train_balanced = list()
y_train_balanced = list()
n = 1895
intervalos = list()
for i in range (62):
    inf = sum(train_counts[:i]) + 1
    inf_mas_1 = sum(train_counts[:i+1])
    sup = min(inf + n, inf_mas_1)
    intervalos.append([inf, sup])
for i in range (62):
    y_train_balanced.extend(y_train_gnb[intervalos[i][0]:intervalos[i][1]])
shuffle(y_train_balanced)
for i in range (len(y_train_balanced)):
    x_train_balanced.append(x_train_gnb[y_train_balanced[i][1]])
    y_train_balanced[i] = y_train_balanced[i][0]
clases62 = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
```

```python
inicio = time.time()
clf_gnb2 = GaussianNB()
clf_gnb2.fit(x_train_balanced, y_train_balanced)
fin = time.time()
print(fin - inicio)
```

```python
y_test_gnb = list(y_test)
test_counts = list()
for i in range (61):
    test_counts.append(y_test_gnb.count(i))
for i in range (len(y_test)):
    y_test_gnb[i] = [y_test[i], i]
y_test_gnb.sort()
x_test_balanced = list()
y_test_balanced = list()
n = 317
intervalos = list()
for i in range (62):
    inf = sum(test_counts[:i]) + 1
    sup = inf + n
    intervalos.append([inf, sup])
for i in range (62):
    y_test_balanced.extend(y_test_gnb[intervalos[i][0]:intervalos[i][1]])
shuffle(y_test_balanced)
for i in range (len(y_test_balanced)):
    x_test_balanced.append(x_test[y_test_balanced[i][1]])
    y_test_balanced[i] = y_test_balanced[i][0]
```
